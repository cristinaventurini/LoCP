```python
import struct

list1 = []
with  open('/mnt/c/Users/ASUS/data/credit_card.dat','rb') as file:
    for line in file:
        list1.append(line)
list1.remove(list1[50])

list2 =[]
for i in range(len(list1)):
    "".join(map(chr, list1[i]))
    list2.append("".join(map(chr, list1[i])))
list3 = []
for i in range(len(list2)): 
    g = list2[i][:-5]
    list3.append(g)
n = 6
list4 = []
for j in range(len(list3)):
    k = [list3[j][i:i+n] for i in range(0, len(list3[j]), n)]
    list4.append(k)
list5 = []   
#cc = [None]*len(list4[i])
for i in range(len(list4)):
    for j in range(len(list4[i])):
        cc = int(list4[i][j], 2)
        list5.append(cc)     
list6 = [list5[x:x+len(list4[0])] for x in range(0, len(list5), len(list4[0]))]
list7 = []
for j in range(len(list6)):
    for i in range(len(list6[i])):
        s = str(list6[j][i])
        list7.append(s)
list8 = [list7[x:x+len(list4[0])] for x in range(0, len(list7), len(list4[0]))]
credit_card = []
for i in range(len(list8)):
    f = "".join(list8[i])
    credit_card.append(f)
print(credit_card[0])
```
